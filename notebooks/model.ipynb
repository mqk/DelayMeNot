## Classification ##

From here on out, I've switched to trying to predict whether a flight will be more than 30 minutes delayed or not. This is a classification problem, and I'm planning to try out a few different techinques: logistic regression, random forests, kNN, etc.

```python
def summarize(Y,Ypred,quiet=False):
    error_rate = 1.0 - float(sum( Ypred == Y )) / len(Y)
    ind = (Y == 0)
    false_positive = 1.0 - float(sum( Ypred[ind] == Y[ind] )) / len(Y[ind])     ## on-time flights classified as late
    false_negative = 1.0 - float(sum( Ypred[~ind] == Y[~ind] )) / len(Y[~ind])  ## late flights classified as on-time
    precision = float(sum( Ypred[~ind] == Y[~ind] )) / sum( Ypred == 1 )
    recall = float(sum( Ypred[~ind] == Y[~ind] )) / sum( ~ind )

    if not quiet:
        print '   Number of flights = %d' % len(Y)
        print '   Actual number of delayed flights = %d (%.1f%%)' % (sum(Y == 1),100.0*sum(Y==1)/len(Y))
        print '   Predicted number of delayed flights = %d (%.1f%%)' % (sum(Ypred == 1),100.0*sum(Ypred==1)/len(Y))
        print '   error rate = %.1f%%' % (100.0*error_rate)
        print '   false positive rate = %.1f%%' % (100.0*false_positive)
        print '   false negative rate = %.1f%%' % (100.0*false_negative)
        print '   precision = %.1f%%' % (100.0*precision)
        print '   recall = %.1f%%' % (100.0*recall)
        print
```

### kNN Classification ###

```python
def knn_classifier(data_train, data_test, train_cols, n_neighbors=5, weights='uniform', quiet=False):

    import time
    from sklearn import neighbors

    knn = neighbors.KNeighborsClassifier(n_neighbors, weights=weights, warn_on_equidistant=False)

    X = data_train[train_cols].values
    Y = np.zeros_like(data_train['ArrDelay'].values)
    Y[data_train['ArrDelay'].values > 30.0] = 1

    knn.fit(X, Y)

    if not quiet:
        print 'Predicting training data...'
    time0 = time.time()
    Y_pred = knn.predict(data_train[train_cols].values)
    dt = time.time() - time0
    if not quiet:
        print '   that took %.1f seconds.\n' % dt

    error_rate = 1.0 - float(sum( Y_pred == Y )) / len(Y)
    ind = (Y == 0)
    false_positive = 1.0 - float(sum( Y_pred[ind] == Y[ind] )) / len(Y[ind])     ## on-time flights classified as late
    false_negative = 1.0 - float(sum( Y_pred[~ind] == Y[~ind] )) / len(Y[~ind])  ## late flights classified as on-time
    precision = float(sum( Y_pred[~ind] == Y[~ind] )) / sum( Y_pred == 1 )
    recall = float(sum( Y_pred[~ind] == Y[~ind] )) / sum( ~ind )

    if not quiet:
        print 'Train sample: '
        print '   Number of flights = %d' % len(Y)
        print '   Actual number of delayed flights = %d (%.1f%%)' % (sum(Y == 1),100.0*sum(Y==1)/len(Y))
        print '   Predicted number of delayed flights = %d (%.1f%%)' % (sum(Y_pred == 1),100.0*sum(Y_pred==1)/len(Y))
        print '   error rate = %.1f%%' % (100.0*error_rate)
        print '   false positive rate = %.1f%%' % (100.0*false_positive)
        print '   false negative rate = %.1f%%' % (100.0*false_negative)
        print '   precision = %.1f%%' % (100.0*precision)
        print '   recall = %.1f%%' % (100.0*recall)
        print

    Ytest = np.zeros_like(data_test['ArrDelay'].values)
    Ytest[data_test['ArrDelay'].values > 30.0] = 1

    if not quiet:
        print 'Predicting test data...'
    time0 = time.time()
    Ytest_pred = knn.predict(data_test[train_cols].values)
    dt = time.time() - time0
    if not quiet:
        print '   that took %.1f seconds.\n' % dt

    error_rate = 1.0 - float(sum( Ytest_pred == Ytest )) / len(Ytest)
    ind = (Ytest == 0)
    false_positive = 1.0 - float(sum( Ytest_pred[ind] == Ytest[ind] )) / len(Ytest[ind])     ## on-time flights classified as late
    false_negative = 1.0 - float(sum( Ytest_pred[~ind] == Ytest[~ind] )) / len(Ytest[~ind])  ## late flights classified as on-time
    precision = float(sum( Ytest_pred[~ind] == Ytest[~ind] )) / sum( Ytest_pred == 1 )
    recall = float(sum( Ytest_pred[~ind] == Ytest[~ind] )) / sum( ~ind )

    if not quiet:
        print 'Test sample: '
        print '   Number of flights = %d' % len(Ytest)
        print '   Actual number of delayed flights = %d (%.1f%%)' % (sum(Ytest == 1),100.0*sum(Ytest==1)/len(Ytest))
        print '   Predicted number of delayed flights = %d (%.1f%%)' % (sum(Ytest_pred == 1),100.0*sum(Ytest_pred==1)/len(Ytest))
        print '   error rate = %.1f%%' % (100.0*error_rate)
        print '   false positive rate = %.1f%%' % (100.0*false_positive)
        print '   false negative rate = %.1f%%' % (100.0*false_negative)
        print '   precision = %.1f%%' % (100.0*precision)
        print '   recall = %.1f%%' % (100.0*recall)

##     return [error_rate, false_positive, false_negative, precision, recall]
    return [error_rate, precision, recall]

```
